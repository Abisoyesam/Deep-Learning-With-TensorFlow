# Non-linearity

```python
# Import libraries

import pandas as pd
import tensorflow as tf
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
```

```python
n_samples = 1000
X, y = make_circles(n_samples=n_samples, random_state=42, noise=0.03)

circles = pd.DataFrame({"X0": X[:,0], "X1": X[:,1], "label": y})
circles.head()
```

```python
# visualize with a plot

plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.RdYlBu)
```

```python
X.shape
```

**Model 1**

```python
# Set random seed
tf.random.set_seed(42)

# 1. Create the model
model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation="linear", input_shape=(2,))
])

# 2. Compile the model
model_1.compile(loss="binary_crossentropy", 
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), 
              metrics=["accuracy"])

# 3. Fit the model
history = model_1.fit(X, y, epochs=100)
```

```python
# Check out our data
plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.RdYlBu)
```

```python
# Check the decision boundary
from plot_func import plot_decision_boundary
plot_decision_boundary(model_1, X, y)
```

**Model 2**
- Change the `activation` to non-linear

```python
# Set random seed
tf.random.set_seed(42)

# 1. Create the model with non-linear activation
model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation='relu', input_shape=(2,))
])

# 2. Compile the model
model_2.compile(loss="binary_crossentropy", 
                optimizer = tf.keras.optimizers.Adam(learning_rate=0.01), 
                metrics=['accuracy'])

# 3. Fit the model
history = model_2.fit(X, y, epochs=100)
```

```python
# Visualize the data
plot_decision_boundary(model_2, X, y)
```

**Model 3**
- Multi layer neural network with `ReLU` activation

```python
# Set random seed
tf.random.set_seed(42)

# 1. Create the model with non-linear activation
model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1)
])

# 2. Compile the model
model_3.compile(loss="binary_crossentropy", 
                optimizer = tf.keras.optimizers.Adam(learning_rate=0.01), 
                metrics=['accuracy'])

# 3. Fit the model
history = model_3.fit(X, y, epochs=100)
```

```python
# Evaluate the model
model_3.evaluate(X, y)
```

```python
# how do the model look
plot_decision_boundary(model_3, X, y)
```

**Model 4**
- Multi layer with `ReLU` activation
- `Sigmoid` activation in the output layer.

```python
# Set random seed
tf.random.set_seed(42)

# 1. Create the model with non-linear activation
model_4 = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# 2. Compile the model
model_4.compile(loss="binary_crossentropy", 
                optimizer = tf.keras.optimizers.Adam(learning_rate=0.01), 
                metrics=['accuracy'])

# 3. Fit the model
history = model_4.fit(X, y, epochs=100)
```

```python
# Evalute the model
model_4.evaluate(X, y)
```

```python
# Visualize the prediction
plot_decision_boundary(model_4, X, y)
```

### Replicate `Sigmoid` function

```python
A = tf.cast(tf.range(-10,10), tf.float32)
A
```

```python
plt.plot(A);
```

$$\sigma (z) = \dfrac{1}{1 + e^{-z}}$$

```python
# Replicate Sigmoid 
def sigmoid(X):
    return 1 / (1 + tf.exp(-X))

# Use the sigmoid function on tensor A
sigmoid(A)
```

```python
# plot the sigmoid tensor
plt.plot(sigmoid(A))
```

### Replicate `ReLU` function

$$
f(x) = \left\{ \begin{array}{11}
        0 & \mbox{for $x < 0$}; \\ 
        x & \mbox{for $x \geq 0$, }\end{array} \right.
$$

```python
def relu(x):
    return tf.maximum(0, x)

# test the relu
relu(A)
```

```python
plt.plot(relu(A))
```

## Evaluating and improving our classification.
- We been training and testing on the same dataset. However, in machine learning, it is wrong to do so. Hence, create a trainig and test set.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Recreate a model to fit on the training data and test on the testing data
# Set random seed
tf.random.set_seed(42)

# 1. Create the model with non-linear activation
model_5 = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# 2. Compile the model
model_5.compile(loss="binary_crossentropy", 
                optimizer = tf.keras.optimizers.Adam(learning_rate=0.01), 
                metrics=['accuracy'])

# 3. Fit the model
history = model_5.fit(X_train, y_train, epochs=50)
```

```python
# Evaluate the model on the test dataset
model_5.evaluate(X_test, y_test)
```

```python
# Plot the decision boundaries for the training and test sets
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.title("Train")
plot_decision_boundary(model_5, X_train, y_train)
plt.subplot(1,2,2)
plt.title("Test")
plot_decision_boundary(model_5, X_test, y_test)
plt.show();
```
