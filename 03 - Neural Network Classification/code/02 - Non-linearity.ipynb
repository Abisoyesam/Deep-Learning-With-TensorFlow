### Replicate `Sigmoid` function

```python
A = tf.cast(tf.range(-10,10), tf.float32)
A
```

```python
plt.plot(A);
```

$$\sigma (z) = \dfrac{1}{1 + e^{-z}}$$

```python
# Replicate Sigmoid 
def sigmoid(X):
    return 1 / (1 + tf.exp(-X))

# Use the sigmoid function on tensor A
sigmoid(A)
```

```python
# plot the sigmoid tensor
plt.plot(sigmoid(A))
```

### Replicate `ReLU` function

$$
f(x) = \left\{ \begin{array}{11}
        0 & \mbox{for $x < 0$}; \\ 
        x & \mbox{for $x \geq 0$, }\end{array} \right.
$$

```python
def relu(x):
    return tf.maximum(0, x)

# test the relu
relu(A)
```

```python
plt.plot(relu(A))
```

## Evaluating and improving our classification.
- We been training and testing on the same dataset. However, in machine learning, it is wrong to do so. Hence, create a trainig and test set.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

**Model 5**
- Splitting to training and testing datasets.
- Evaluating on testing datasets

```python
# Recreate a model to fit on the training data and test on the testing data
# Set random seed
tf.random.set_seed(42)

# 1. Create the model with non-linear activation
model_5 = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# 2. Compile the model
model_5.compile(loss="binary_crossentropy", 
                optimizer = tf.keras.optimizers.Adam(learning_rate=0.01), 
                metrics=['accuracy'])

# 3. Fit the model
history = model_5.fit(X_train, y_train, epochs=50)
```

```python
# 4. Evaluate the model on the test dataset
model_5.evaluate(X_test, y_test)
```

```python
# Plot the decision boundaries for the training and test sets
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.title("Train")
plot_decision_boundary(model_5, X_train, y_train)
plt.subplot(1,2,2)
plt.title("Test")
plot_decision_boundary(model_5, X_test, y_test)
plt.show();
```

## Finding the best learning rate

To find the ideal learning rate (the learning rate where the loss decreases the most during training), you the following steps:

1. A learning rate **Callback**: You can think of a callback as an extra piece of functionality you can add to the model while training.
2. Another model (we could use the same one as above)
3. A modified loss curves plot.

**Model 6**

```python
# Set random seed
tf.random.set_seed(42)

# 1. Create the model with non-linear activation
model_6 = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# 2. Compile the model
model_6.compile(loss="binary_crossentropy", 
                optimizer = "adam", 
                metrics=['accuracy'])

# 3. Create a learning callback
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10 ** (epoch/20))

# 4. Fit the model
history_6 = model_6.fit(X_train, y_train, epochs=100, callbacks=[lr_scheduler])
```

```python
pd.DataFrame(history_6.history).plot(figsize=(10,7), xlabel="epochs")
```

```python
# Plot the learning rate versus the loss
lrs = 1e-4 * (10 ** (tf.range(100)/20))
lrs
```

```python
plt.figure(figsize=(10,7))
plt.semilogx(lrs, history_6.history["loss"])
plt.xlabel("Learning Rate")
plt.ylabel("Loss")
plt.title("Learning Rate Vs Loss")
```
